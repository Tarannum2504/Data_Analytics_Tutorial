# Data Structures

1. LIST:
Ordered collection,mutable, allows duplicate

```python
#Core Properties
# 1. Ordered
nums = [10, 20, 30]
print(nums[0])  # 10

#2. Mutable- value can be changed
nums[0] = 100   

# 3. Duplicates Allowed
nums = [1, 2, 2, 3]

# 4. Dynamic Size 
nums.append(4)


# important methods
nums.append(4)      # add at end
print(nums)
nums.insert(1, 10) # insert at index
print(nums)
nums.remove(2)    # remove first occurrence
print(nums)
nums.pop()        # remove last
print(nums)
nums.sort()        # sort list
print(nums)

#Important Tricks
# Copy Issue (VERY COMMON BUG)
a = [1,2,3]
b = a
b[0] = 100

print(a)  # [100,2,3] Same reference (not copy)

# correct way to copy
b = a.copy()

# appending 1 list after another 
a= [1,2,3,4]
b=[5,6,7]
a.extend(b)
print("after extending a become:",a)
a.append(b)
print("after appending b , a becomes:",a)

# python min() on a list of strings compares them Alphabetically or lexicographically
cities = ["Delhi","Mumbai","Chennai"]   # wipro interview question
print(min(cities))

```

2. TUPLES:
Ordered collection, immutable



```python
t = (1, 2, 3)

#Core Properties
# 1. ordered
t = (10, 20, 30)
print(t[0])  # 10

# 2. Immutable - You cannot change, add, or delete elements
t = (1, 2, 3)
t[0] = 10   #  Error

# 3. Allowss duplicate
t2 = (1, 2, 2, 3)

# 4. Can Store Mixed Data Types
t3 = (1, "hello", 3.5, True)

# ⚠️ Important Trick
# Single Element Tuple
t = (5)     #  Not a tuple
t = (5,)    #  Tuple     ( Comma is important, not parentheses)

# Tuple Operations
# 1. Access
t = (10, 20, 30)
print(t[1])  # 20

# 2. Slicing   tuple_name[ start: end-1]
print(t[0:2])  # (10, 20)

# 3. Concatenation
t1 = (1, 2)
t2 = (3, 4)
print(t1 + t2)  # (1,2,3,4)

# 4. Repetition
print((1, 2) * 2)  # (1,2,1,2)

# Packing & Unpacking
# packing 
t = 1, 2, 3   # automatically becomes tuple

# Unpacking
a, b, c = (1, 2, 3)
print(a)  # 1

# Functions Returning Multiple Values
def calc():
    return 10, 20

x, y = calc()    # Actually returns a tuple

# Why Tuple is Important?
# 1. Faster than list as it uses less memory and is optimized
# 2. it is safe (immutable) -> no accidental changes 
# 3. can be used as a dictionary key
d = {(1,2): "point"}  
# Lists cannot be keys

# ⚠️ Mutable Inside Tuple -> Tuple is immutable, but inside mutable objects can change
t4 = (1, [2, 3])
t4[1][0] = 100   # allowed


```

3. SETS:
Unordered collection of unique elements

```python
s={1, 2,2, 3} # {1, 2, 3} duplicates removed, order not guaranteed

#core properties
# 1. Unordered - No indexing, no slicing    
# 2. Unique Elements - Duplicates removed
# 3. Mutable - You can add/remove elements, but not change existing ones
s.add(4)      # add element
s.remove(2)   # remove element
# 4. fast lookup due to hashing
print(3 in s)  # membership test is O(1)

# Set Operations
s1 = {1, 2, 3}
s2 = {3, 4, 5}
print(s1.union(s2))         # {1,2,3,4,5} 
print(s1 | s2)           # {1,2,3,4,5}  (union operator)
print(s1.intersection(s2))  # {3}   
print(s1 & s2)           # {3}  (intersection operator)
print(s1.difference(s2))    # {1,2}
print(s1 - s2)           # {1,2}  (difference operator)
print(s1.symmetric_difference(s2))  # {1,2,4,5}
print(s1 ^ s2)           # {1,2,4,5}  (symmetric difference operator)

# Important Tricks
# 1. Empty Set
s = {}  # This creates an empty dictionary, not a set
s = set()  # Correct way to create an empty set

# 2. Set Comprehension
squares = {x**2 for x in range(5)}  # {0,1,4,9,16}

# 3. Mutable Elements Not Allowed
s = {1, 2, [3, 4]}  # Error: unhashable type: 'list'
```

4. DICTIONARIES : 
Unordered collection of key-value pairs

```python
d = {"name": "Alice", "age": 30}

# Core Properties
# 1. Unordered - No indexing, no slicing
# 2. Key-Value Pairs - Each key maps to a value
print(d["name"])  # Alice

# 3. Mutable - You can change values, add/remove key-value pairs
d["age"] = 31   # update value
d["city"] = "NY" # add new key-value pair

# 4. Keys Must Be Unique - Duplicate keys will overwrite previous value
d = {"a": 1, "a": 2}  # {'a': 2}

# 5. Keys Must Be Immutable - Lists cannot be keys, but tuples can
d = {[1, 2]: "value"}  # Error: unhashable type
d = {(1, 2): "value"}  # Valid

# Dictionary Operations
d = {"a": 1, "b": 2, "c": 3}
print(d.keys())    # dict_keys(['a', 'b', 'c'])
print(d.values())  # dict_values([1, 2, 3])
print(d.items())   # dict_items([('a', 1), ('b', 2), ('c', 3)])
print(d.get("b"))  # 2 (returns None if key not found)
print(d.get("z", "default"))  # default (returns default if key not found)
d.pop("a")  # removes key 'a' and returns its value
print(d)  # {'b': 2, 'c': 3}
d.update({"d": 4, "e": 5})  # add multiple key-value pairs

# Important Tricks
# 1. Dictionary Comprehension
squares = {x: x**2 for x in range(5)}  # {0:0, 1:1, 2:4, 3:9, 4:16}

# safe access v/s direct access
d = {"a": 1, "b": 2}    
print(d["c"])  # KeyError
print(d.get("c"))  # None

# 2. Iterating Over Dictionary
d = {"a": 1, "b": 2, "c": 3}
for key in d:  # iterates over keys
    print(key)

for value in d.values():  # iterates over values
    print(value)

for key, value in d.items():  # iterates over key-value pairs
    print(key, value)

```

# Error Handling:
Used to prevent program crashes.

#Core structure

try:

        #risky code

except:

        #handle error

Core Properties:

try:
Code that may fail.

except:
Handles failure.

else:
Runs if no error occurs.

finally:
Runs always.

```python
#Example
try:
    print(10/2)

except ZeroDivisionError:
    print("Error")

else:
    print("Success")

finally:
    print("Finished")

# Common Exceptions
# 1. ValueError
int("abc")

# 2. TypeError
"5"+5

# 3. IndexError
nums=[1,2]
print(nums[5])

# 4. KeyError
d={"a":1}
print(d["b"])

# 5. FileNotFoundError
open("abc.txt")

# Important Tricks
# Multiple Exceptions
except (ValueError,
        TypeError):

# Raise Exception
raise ValueError(
"Invalid Salary"
)

# Pitfalls
# Catching Everything

except:
    pass

# Bad practice.
# Hides bugs.

# Ignoring Error Message
except:
    print("Error")

# Better:

except Exception as e:
    print(e)

# Why Important?

# Real datasets contain:

# Missing values
# Wrong types
# Corrupt files
# Invalid records

# Without error handling, analytics pipelines fail.
```

# Comprehensions:
Shorter way of creating collections.

Core Properties
1. Faster: 
    Generally faster than loops.

2. Cleaner: 
    Less code.

3. More Pythonic:

```python
## List Comprehension
# Normal Way
squares=[]

for i in range(5):
    squares.append(i*i)

# Comprehension
squares=[i*i for i in range(5)]

# Conditional Comprehension
even=[x for x in range(10)
      if x%2==0]

# If Else Comprehension
result=[
"Even" if x%2==0 else "Odd"
for x in range(5)
]


# Dictionary Comprehension
squares={
x:x*x
for x in range(5)
}

# Set Comprehension
unique={
x for x in [1,2,2,3]
}

# Important Tricks
# Nested Comprehension
matrix = [
[x*y for y in range(3)]
for x in range(3)
]

# Pitfalls
# Too Much Nesting

result=[
x*y
for x in range(10)
if x%2==0
for y in range(10)
if y%3==0
]

# Hard to read.
# Sometimes loop is better.

# Why Important?

# Very common in:

# Data Cleaning
# Feature Engineering
# Filtering
# Transformations
```
